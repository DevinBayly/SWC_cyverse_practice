````r
vec_country <- c("China","Czech Republic","Cuba")
gap.in %>%
filter(country %in% vec_country) %>%
group_by(country) -> country_data
````

````r
for (name in vec_country) {
   current_country= filter(country_data,country == name)
    plot(current_country$gdpPercap ~ current_country$year,main = name,xlab = 'year',ylab = 'gdp')
}
````
